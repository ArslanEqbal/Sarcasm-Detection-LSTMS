```python
import nltk
```

```python
import re
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
corpus = []
```

```python
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['headline'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word)for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Dropout
```

```python
voc_size = 5000
```

```python
one_hot_representation = [one_hot(words,voc_size)for words in corpus]
```

```python
one_hot_representation
```

```python
sentence_length = 20
embedded_values = pad_sequences(one_hot_representation , padding = 'post' , maxlen = sentence_length)
```

```python
embedded_values
```

```python
embedded_values[1]
```

```python
len(embedded_values)
```

## MODEL

```python
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size , embedding_vector_features , input_length = sentence_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam' , metrics = ['accuracy'])
model.summary()
```

```python
import numpy as np

x_final = np.array(embedded_values)
y_final = np.array(y)
```

```python
x_final.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test , y_train , y_test = train_test_split(x_final , y_final ,test_size = 0.3 ,random_state =10)
```

## MODEL FITTING

```python
model.fit(x_train,y_train,validation_data = (x_test , y_test) , epochs = 20 , batch_size = 32)
```
